```python
import csv
import random
import operator
from typing import List, Tuple
import math

def load_iris_dataset(filename: str) -> List[List]:
    dataset = []
    with open(filename) as f:
        reader = csv.reader(f)
        next(reader)  # Skip header
        for row in reader:
            dataset.append([float(x) for x in row[:4]] + [row[4]])
    return dataset

def split_dataset(dataset: List, split_ratio: float) -> Tuple[List, List]:
    train = []
    test = []
    for row in dataset:
        bucket = test if random.random() < split_ratio else train
        bucket.append(row)
    return train, test

def calculate_distance(instance1: List, instance2: List) -> float:
    return math.sqrt(sum((x - y) ** 2 for x, y in zip(instance1[:-1], instance2[:-1])))

def get_nearest_neighbors(training_set: List, sample: List, k: int) -> List:
    distances = [(calculate_distance(sample, neighbor), neighbor) for neighbor in training_set]
    distances.sort(key=lambda x: x[0])
    neighbors = [neighbor for _, neighbor in distances[:k]]
    return neighbors

def customGetResponse(neighbors: List) -> str:
    classVotes = {}
    for neighbor in neighbors:
        response = neighbor[-1]
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1

    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
    return sortedVotes[0][0]

def get_class_mapping(dataset: List) -> dict:
    class_column = -1
    classes = set(row[class_column] for row in dataset)
    class_mapping = {cls: cls for cls in classes}  # Map class names directly
    return class_mapping

def print_readable_prediction(test_sample: List, prediction: str, class_mapping: dict) -> None:
    class_label = test_sample[-1]
    readable_prediction = class_mapping.get(prediction, prediction)
    readable_class_label = class_mapping.get(class_label, class_label)
    print(f"Prediction for {test_sample[:-1]}: {readable_prediction} (Actual: {readable_class_label})")

if __name__ == "__main__":
    dataset = load_iris_dataset("iris.csv")
    training_set, test_set = split_dataset(dataset, 0.66)

    class_mapping = get_class_mapping(dataset)

    for test_sample in test_set:
        k_value = 3  # Adjust the value of k
        neighbors = get_nearest_neighbors(training_set, test_sample, k_value)
        prediction = customGetResponse(neighbors)

        print_readable_prediction(test_sample, prediction, class_mapping)

```
